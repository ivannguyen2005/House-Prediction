```python
df1 = df.drop(['city', 'zip_code','prev_sold_date','brokered_by','street'], axis=1)
df1.sample(5)
```

```python
print(df1.isna().sum())
```

```python
#Filling missing value
df1['price'] = df1['price'].fillna(df1['price'].mean())
df1['bed'] = df1['bed'].fillna(df1['bed'].median())
df1['bath'] = df1['bath'].fillna(df1['bath'].median())
df1['acre_lot'] = df1['acre_lot'].fillna(df1['acre_lot'].median())
df1['house_size'] = df1['house_size'].fillna(df1['house_size'].median())
df1 = df1.dropna(subset=['state'])
```

```python
print(df1["status"].value_counts())
```

```python
#transform status 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['status_encoded'] = le.fit_transform(df1['status'])
df1
```

```python
#transform states 
print(df1["state"].value_counts())
df1 = df1.drop(df1[df1["state"].map(df1["state"].value_counts()) < 50]["state"].index)
df1['state_encoded'] = le.fit_transform(df1['state'])
df1= df1.drop("status", axis =1)
df1= df1.drop("state", axis =1)
df1
```

```python
#check duplicates and delete 
num_duplicates = df1.duplicated().sum()
df_cleaned = df1.drop_duplicates()
```

```python
#checking data again
df_cleaned.info()
df_cleaned.describe()
```

```python
# remove outliers for prices
from scipy import stats
numeric_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Calculate Z-scores for all numeric columns
z_scores = df_cleaned[numeric_columns].apply(stats.zscore)

# Filter rows: Keep only rows with all Z-scores within the threshold (-3 to 3)
df_no_outliers = df_cleaned[(z_scores.abs() <= 3).all(axis=1)]
df_no_outliers
```

```python
#create new feature
df_no_outliers.loc[:, 'total_rooms'] = df_no_outliers['bed'] + df_no_outliers['bath']
df_no_outliers.loc[:, 'room_density'] = df_no_outliers['total_rooms'] / df_no_outliers['house_size']
```
